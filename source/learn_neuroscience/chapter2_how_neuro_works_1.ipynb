# 静息态的神经元膜

神经信号是如何传递的？在细胞外液中，含盐的环境下，要想让轴突长距离传输，同时尽量不损失信号，就要采用特殊的手段。在上一章我们见到了胶质细胞形成的髓鞘增加了导电性。而在这一章，我们主要着眼在信号本身——这一被称之为**动作电位（action potential）**的信号：采用固定大小、持续时间的稳定不衰减传输背后的生理机制。

动作电位是在**可兴奋性膜（excitable membrane）**上进行传输的。此膜在神经细胞和肌细胞上都有分布，当可兴奋性膜从静息态转变为动作态时，其转变时间很短，大约1ms。这两种状态称之为**resting membrane potential(静息膜电位)**和动作膜电位。

## 化学特性

excitable membrane 的功能来自于三个组件：膜内外的盐溶液、膜以及跨膜的蛋白质。

对于膜内外的盐溶液而言，也就是细胞质和细胞外液，最主要的成分是**水（H2O）**。水是一种由两个氧原子和一个氢原子由极性共价键组成的，这一结构导致了氧原子带负电，氢原子带正电。这种极性分子容易溶解像离子键构成的盐、其他极性共价键构成的分子，但是不容易溶解那些电荷分布均衡的油和脂。

除了水之外，其次重要的东西是**离子（ion）**。像盐溶液这种离子，由Na+和Cl-离子键组成，这一键没有水的极性共价键强，因此容易被拆散，所以离子容易溶于水。其中Na+离子和水的O-端结合，被其包围，Cl-例子和水的H+端结合，被其包围。

神经传递发生在膜上，众所周知，膜是**磷脂双分子膜**，两层结构。产生这种情况的原因是：磷脂分子的磷酸基团含有一个极性的磷三氧结构，因此容易溶于水。此外，其含有碳氢键则结构稳定，不容易溶于水。因此，磷酸双分子层采取背靠背的方式，水亲和性高的部分暴露在细胞质和细胞外液中，而亲和性低的碳-氢键则处于中间。

**蛋白质**由氨基酸构成。其中氨基酸的结构是：一个α碳，一个氢原子、一个氨基和一个羧基，一个R基团。其中氨基和羧基组成肽键（peptide bond），而这一条由多个氨基酸分子组合的结构称之为多肽（polypeptide）。对于氨基酸而言，其亲水和疏水特性来自于R基团的构造。当氨基酸折叠形成亚基：二级结构是双螺旋，三级结构是3D空间折叠，最后形成的四级结构组成了蛋白质。因为蛋白质是由氨基酸折叠构成的，因此其可以在部分疏水、部分亲水。这完全取决于蛋白质所使用的氨基酸类型以及其折叠方式。

镶嵌在神经元细胞和肌细胞的可兴奋性膜上的一些蛋白质，称之为**通道蛋白**。这些蛋白构成了**门控通道（gating）**，具有离子选择性。其能够镶嵌在膜上的一个原因是：这些通道蛋白镶嵌部位的氨基酸的R基团是疏水的，因此和膜上的C-H端相似，而暴露在细胞质和细胞外液中的其他部分则是亲水的，因此和膜上的P-O端类似，相似的特性容易让这种镶嵌保持稳定状态。

膜上的蛋白质除了形成离子通道之外，还能够形成**离子泵（ion dump）**。离子泵是一种酶，其消耗ATP来将钠离子和钾离子运送出细胞膜外，其对于神经信号的传递起到决定作用。

## 离子的运动

离子的运动由于两种原因，其一是浓度主动扩散，离子将会从高浓度向低浓度进行扩散**（conctration gradient）**，直到平衡为止。其二是电荷迁移平衡。正负电荷总要达到一个均等的状态。

电的产生需要以下条件：电导（electrical conductance），电导是电阻的倒数，表示电子能够通过的程度。电压指的是阳极和阴极之间的电荷差异，其定义为施加在带电粒子上的力。I = gV，其中电流需要电导，如果 g=0，那么即便有再高的电压，也不会产生电流。

**静息膜电位产生的离子基础**：**静息膜电位（membrane potential）** 指的是在任意情况下平均的膜两侧的电位差，用 Vm 表示。典型的 Vm = -65 mV。静息膜电位指的是这样一种平均：膜上含有各个离子的通道，离子因为浓度和电特性在两侧进行迁移，同时离子泵也在起着作用，综合这些的一个平均电位差是 -65 mV，也就是细胞膜内测的电位低，含有更多的负电荷，细胞膜外侧的电位高，含有更多的正电荷。

**平衡电位（equlibrium potential）** 是对于某一离子而言的，其定义如下：如果大量的注入离子 K+ 和 A-，同时关闭所有离子通道和离子泵，这时因为两侧电荷平衡，因此 Vm = 0 mV。而现在如果插入 K+ 的离子通道，因为浓度原因，K+ 离子将大量的流向外部，而当这一过程发生的时候，在细胞内部含有越来越多的 A-，导致了 Vm 的升高（负值），因此由于电的作用，细胞内部将捕获更多的 K+ 以维持电荷的平衡。当这两种力互相平衡的时候，K+ 离子的运动停止，这个时候的电位差称之为 Eion，简称平衡电位。对于 K+ 而言，是 -80mV。

由于磷酸双分子层非常的薄，因此，细胞内的负电荷和细胞外的正电荷往往在此处产生吸引，这导致了膜能够具有一定程度的电容。Eion-Vm 指的是**离子驱动力（ion diriving force）**，它决定了对于某种离子在细胞膜内外进行移动的方向。

在上述的例子中，如果细胞外的 Na+ 离子更多，并且具有通路，那么平衡时将会导致细胞内具有正电荷。根据 Nernst 方程，可以通过离子的电荷、温度、细胞内外离子浓度比值，能够计算出离子的平衡电位值。如下所示，其中 R 表示气体常数，T 表示绝对温度，Z 表示离子电荷，F 表示法拉第常数。后面的分数指的是外部和内部的离子浓度比。

$$E_{ion}=2.303\frac{RT}{ZF}\frac{[ion]_o}{[ion]_i}$$

使用 Nernst 方程不需要知道通透率或者离子电导。如果是负电荷，那么E 为负，反之为正。

**离子的跨膜分布** 离子除了由于浓度梯度移动，以及电和浓度的平衡以外，离子泵也起到了关键的作用。Na-K泵总是将细胞膜内的Na+和细胞膜外的K+进行交换，在大脑所有的ATP消耗中，Na-K泵酶的能量消耗占据了70%。此外还有一种酶称之为Ca泵，其将细胞内的Ca2+运送到细胞外。离子泵的工作导致了某些离子在细胞膜内外的分布具有某种稳定性：K+离子总是膜内的浓度高，Na+和Ca2+离子总是膜外的浓度高。

考虑到真实的情况，细胞膜运行Na+和K+的通透，那么，对于Vm而言，应当是Ek和Ena的一个平均，但是，因为通透性不同，K+离子的通透性更高，而Na+的通透性更低，因此，Vm更接近Ek。考虑带有不同的通透性的 Vm 的计算通过 Goldman 方程可以计算。

细胞膜上的蛋白质是如何做到对于K+离子有更高的通透性的？这取决于一个称作 pore loop 的结构。如果此处的某个氨基酸被破坏，那么将会导致严重的后果。当神经元电位的负值减小，这将严重影响神经元的信号传输。实际上，大量注射 KCl 可以导致神经细胞去极化（depolarization）。大脑中的血脑屏障会限制K+进入大脑的细胞外体液中，神经胶质细胞，特别是星状胶质细胞，会利用细胞质膜上的K+泵吸收大量的K+，导致其自身去极化，从而保护了大脑中的神经元正常的信号传输。

但是对于肌肉细胞而言，没有这样的屏障。显而易见的就是安乐死时的注射致死，心肌细胞不再接受信号，导致心脏停止跳动。
