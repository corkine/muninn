## 1.2 Seeing Distance with One Eye

定义距离（Distance）指的是一种单眼知觉（monocularly），其表示我们能够感知到场景中的对象的距离。（Distance只涉及单眼知觉，没有双眼差异）

定义深度（Depth）表示stereopisis（立体知觉），指的是一种由双眼知觉（binocularly）所调节的对深度的认识。

<img src="seeingspace13.png" width="50%" height="50%" align="center">

### 1.2.1 Cues

常见的单眼视觉线索有：

1、最重要的：透视——一种几何关系，近大远小

2、遮挡：被遮挡的物体一般位于远处

3、空气透视：由于蓝天反射蓝色光纤，还有空气过滤掉部分远处物体光线，因此远处物体看起来更蓝、更模糊

4、运动错觉：近处的物体运动更快，远处更慢。

### 1.2.2 Distance “illustion” Explanation

除此之外，还有一个非常有意思的问题：

<img src="seeingspace14.png" width="50%" height="50%" align="center">

当我们俯视和仰视物体时，我们估计物体和我们的距离会发生变化。对于俯视而言，估计距离总是小于实际距离，并且其差异逐渐随着俯视角度增大而减小。（低估）。也就是说，当越来越接近平视的时候，估计准确度变得更低。

在平视的时候更有趣，当距离较近的时候，我们估计到的距离更远。但当距离较远的时候，我们估计的距离更近。

在仰视的时候，情况和俯视相反，我们估计到的距离相比真实距离更远。（高估）

<img src="seeingspace15.png" width="70%" height="50%" align="center">

同样的，使用laser range finder进行研究。不同颜色表示不同的可能性，可以看到，最高可能发生在较近的俯视，在这里我们倾向于低估，准确性较高。而距离较远的俯视可能性很低，在这里我们估计的准确性较差，但依旧倾向于低估。而中等概率发生在仰视和较近的平视，在这里我们倾向于高估。

我的解释（一种可能性）：对于俯视，我们很容易遇到别的物体，所以低估有利于警戒。而远处的物体，虽然其出现概率较低，但是可能有奔跑速度很快的哺乳动物，我们自己也可以跑得很快，所以依旧是低估，由于距离增多，判断的线索较少或者因为在这个地方面对的生存压力最大，因此估计准确性最低（被严重加权）。而对于样式，我们很少遇到物体，所以倾向于高估以减少大脑压力，对于较近的平视和仰视类似。对于较远的平视，其遇到物体的概率也不高，但是估计面临较大的生存压力，所以也倾向于低估。

所以我认为，可能是进化的压力导致的对于俯视和仰视、平视的这种出现概率×重要性加权的视觉距离估计响应模式。

## 1.3 Seeing Depth with Two Eyes (Stereopsis)

### 1.3.1 Basic Stereopsis

实际上，很多动物都没有深度视觉，但是它们依然生活的很好，而那些只有一个眼能够工作的人生活上也没有太大的问题。但是，拿两个笔，试着将笔头怼在一起，你会发现双眼比单眼容易的多，这就是深度视觉的作用。深度视觉的物理原理非常简单，就是因为6.5cm远的双眼距离导致成像的不同，这导致了视网膜上存在这双眼时差，和之前一样，双眼视差是一个不可逆问题。双眼时差如何被转换、通信如何被加工，在神经层面上还有各种现象层面上都没有很好的被解决。

<img src="seeingspace16.png" width="70%" height="50%" align="center">


Wheatstone是最早研究双眼视觉的科学家。其观察到，左右眼在注视同一个点的情况下，视野中的图像是不同的。造成这种结果的原因是——当我们在注视某个东西时，对于非注视点，其在固定不动的双眼中所成的像的位置存在差异，这造成了视觉上左右眼的不同。stereopsis toy就是利用了这种原理，给左右眼呈现不同的图案，这些图案对应视网膜上的左右眼不同的像，然后大脑就像感知到立体物体一样感知到这些差异。

Wheatstone’s “pseudoscope” 幻视镜是一种很特别的东西，其在左右眼经过的路线上各加上了一个三棱镜，导致左眼能看到右眼的线索，右眼能看到左眼的线索。像是倒的。

<img src="seeingspace17.png" height="30%" align="center">

如下图所示，我们认为双眼视觉的一个解释在于，对于圆弧⌒上的点，其具有和视网膜相同的距离，我们认为其深度相同。而当视觉注视在黑点的时候，近处的物体透过当前固定的眼球在左右眼形成不同位置的像，一个为p，一个为q，p和q的差异就是深度知觉产生的物理基础。远处点类似，n和m的差异就是远处的深度知觉。

需要注意，深度知觉只有30m左右，再大的距离，双眼看到的像的差异过小，就很难产生深度知觉。

### 1.3.2 The Inverse Problem

真实世界中的像的特征包括其几何特征我们并不清楚，并且不能够通过任何的探测或者算法来了解。通信问题是立体知觉中不可知的外在刺激的表现。也就是说，我们不清楚外界的深度信息，但是我们有一个双眼的视差，pq和nm，我们只知道这个差别，任何外在刺激能够相同表现这个差别的，即便其位置不同，我们也无法计算和知晓。这就是不可逆问题在双眼视觉中的表述。

### 1.3.3 The Correspondence Problem # Behavior

**Julesz Random Dots Stereograms实验**

![](seeingspace23.png)

如上所示是Béla Julesz的Random Dot Stereograms实验。这个实验是为了解决The Correspondence Problem存在的。通信问题指的是，左右眼如何将视觉差异合并在一起？

在这个试验中，要求被试双眼分别观察左右两个随机点，当在右眼看到一个物体时，在左眼会期望有一个物体的近似的视觉，但是左边仅仅是随机点，所以我们就人为的建构出一个稍微有差异的像，这导致了我们在随机点上看到了立体的物体，可能是一个凹陷的方形或者是一个凹陷的圆。

**Sherrington闪光融合实验**

![](seeingspace21.png)

如上图所示的是Charles Sherrington所做的实验，左右眼观察不同的闪光频率。我们猜测，同步左右眼导致闪光频率不会变化，而异步则会导致闪光频率发生变化，因为间隔减小了。但是实际没有出现变化，异步情况下，闪光融合频率和同步一样。这导致了双眼视觉如何融合的这个问题变得十分棘手。

**Binocular Rivalry**

这个问题的关键在于我们只能看到遮罩之内的棍子，因此我们不清楚真实的物体运动方向。而不同的现实世界运动方向在对应视网膜方向时，其长度和速度都发生了变化，因此，我们得到的结果如下图，杏仁状为出现在各个方向上的可能性，其中垂直于棍子方向的运动可能性最大。同样的原因——因为我们看到的方向只是其投影在视网膜上的分量。

![](seeingmotion7.png)
![](seeingmotion8.png)

为什么是垂直呢？因为垂直于物体的运动方向比其他运动方向更容易满足孔径。这是人类的经验所导致的。而对于垂直遮罩也是的，因为重力的作用，a有更大概率出现。

什么叫做满足孔径？这里指的是，穿过孔径同时没有露出任何的端口。因为垂直的运动方向最常在孔径中发生这种现象，而对于别的方向，比如右图蓝色的线，必须有更长的线段才能够出现这种现象，因此，垂直方向比别的方向更容易出现这种现象。同样的，对于垂直孔径，垂直运动的线出现这种不露端口的线概率更大，更容易满足孔径，所以我们很大概率这样感知——垂直运动。

看起来我们有一种默认反应的模型，来应对随机不确定的变化（不确定线段长度）。

如上图所示，对于麻醉的和人有类似视觉特征的哺乳动物-猫进行实验，记录来自V1的某个细胞的tuning反应，这个细胞只能接受刺激的某个范围，因此A中呈现的刺激要在这个范围内。B是测试条件，看到这个细胞对于方向有所偏好，有些方向根本不激活，而有些方向激活程度也不同。

**结论：** 神经活动确实代表了图像的特征

但是问题是，**图像的特征（Stimulate）并不代表真实的物理世界（Phy World）的刺激**。因此，特征检测可以看作是这些特征生理处理的表示，但是前期如何将3D转换成2D并且使用经验处理和猜测2D以还原3D，这个重要的过程是缺失的。特征检测可以看作是视觉知觉的光学-神经电信号转换（特定神经元对应特定刺激光学特征，并且，中间可能进行了其余的处理，比如经验性的加权和猜测，我的猜测是，在能量形式转换的过程中并非线性转换，而是一个基于统计和概率的转换）。

特征检测论中天真的认为图像对应现实世界我们看到的物理刺激是错误的。



### 3.1.2 推导

赫尔姆霍兹认为，人类的视觉光学系统很差，但是视网膜得到的信息是完整的（错误）。因此，我们会下意识主观根据经验来获取更多的信息（思路很对）。

其中贝叶斯定律， $P(A\mid B) = P(B\mid A)P(A)$ A代表stimulus，B代表物理世界，可以求得在真实世界情况下发生的感知刺激的概率。但是问题是，我们不可能知道 $P(B\mid A)$ ，因为我们的视网膜-大脑永远不知道真实世界。

### 3.1.3 编码

根据Shannon的信息压缩，Horace Barlow 认为，外界信息过多，我们可能也有一个对信息进行压缩并且解码的过程，就像刺激在神经元的树突——轴突之间传递一样，刺激有不同水平，但是只有0和1两种结果。但是信息压缩和解码就像电话线通信，其并不涉及我们看到的内容，或者如何看到的，其是关于我们如何高效的传输信息的过程。

### 3.1.4 视觉作为Inverse问题的对抗

古怪的观察方式的原因是因为我们视觉的策略导致的，我们并非对外界刺激进行特征感知，而是对其可能性和正确性进行估计。之后通过特征检测论中发现的神经细胞的转换，加权人类经验进行可能的猜测和转换，最后在皮层进行电信号处理。

## 3.2 前瞻

对于膝跳反射而言，事情并非向我们想象的那么简单。有两个很有意思的现象：
1、当被试被敲打多次，最近一次敲打，即便没有敲击到，被试也会做出反应
2、高层系统，比如皮层，可以阻断反射。
比如对于医院的医生，如果怀疑一个病人的膝跳反应是否受损，要求被试紧绷肌肉，如果在这种状况下进行检查没有反应才算没有反应。而如果被试非常放松，可能就不会有反应。这说明了这种经典性条件反射存在非常复杂的过程。

大脑和计算机（不包含统计的典型算法）的比喻是错误的。相反，和大脑比较相近的是机器学习，比如遗传算法等开放式的非监督学习。但是对我而言，我更认同监督学习中的神经网络cnn算法，这可能是在一个较短时间范畴内有用的（而非监督学习则是在很长的时间范畴内有用的），并且可能和神经元工作类似的算法。

cnn的作用可能在于：我们已经有了进化的经验。那么，这是一个监督学习的例子，和神经元工作方式类似，cnn猜测视觉输入，并且有选择的将其光信号转换成为电信号。

# 4. Ponzo effect 实验设计
我认为Ponzo effect的视觉错觉非常有趣，我认为这可能是我们试图2D平面上还原3D物理空间的刺激时犯下的一个错误。

![](tmp1.png)


我的解释如下：

众所周知，我们的视网膜是一个2D结构，因此，物理世界的3D刺激不能够完全的被我们的视觉感受器所接受，因此出现了很多的视觉错觉现象。我们的眼睛感知距离的时候，主要依靠的是单眼线索。在这些线索中最重要的一个就是近处的东西看起来大，远处的东西看起来小，这被称之为透视。

思考一下不可反转的视觉感知问题。我们的视网膜感受到的仅仅是一个2D平面，因此主要依靠透视来还原3D物理世界的刺激。而这总会出错，因为我们所接受到的信息是不完全的，所以也没有一种数学的算法进行精确地还原。

在之前的课程中，Dale教授向我们展示了大脑对于这种不可逆问题的处理办法：根据先天的经验进行猜测，这也是他认为所有视觉错觉问题的源头。我认为这就像是一种统计而非数学的解决方法，可能在大脑中的丘脑或者初级视觉皮层的某处，神经元通过“神经网络”算法对输入的信息按照已有的模型进行加权，然后输出一个修正过的结果。这样的计算方法并不是精确的，而是根据之前的经验学习到的，因此也被归为机器学习领域，我认为这和大脑处理视觉刺激有非常相似的地方。

回到Ponzo问题上来。我的假设是，那两条不平行的黑色竖线，造成了我们对于透视关系的感知。因此，就像我们对色彩、明度、物体大小的感知一样，它们收到周围环境的影响。我认为对于两条“实际”相等的红色线段的判断而言，这两条黑色的不平行的竖线造成了一种“模拟的空间知觉”，让我们在2D的图画上看到了3D的物理世界，正如其经常性的对于外界的3D线索投射到视网膜上的2D图画所做的一样，反正我们看到的是2D图像，但是，大脑根据以往的经验认为，我们看到的是3D图像，它的理由是透视，因为那两条黑色的线段看起来就像我们经常看远方笔直的公路一样，这在我们整个进化过程中是经常看到的（毕竟我们进化的经验很大一部分时间并不是注视着纸或者iPad或者电脑屏幕）。

因此，根据黑色竖线这个线索，大脑猜测我们在看3D图案（它的确无法区分真实的3D还有电脑屏幕上的平面，因为它所知道的仅仅是视网膜上的2D信息）。

如果我们的知觉认为我们在观察一个3D场景的话，自然会认为远方的物体小而近处的物体大。而对于实际相等的线段而言，其位置如果靠近远端，那么其会被加权，并且认为比真实的要更大。如果其靠近近端，那么其会被降低权重，并且被认为比真实的要更小。被认为靠近近处还是远处取决于我们的透视线索，也就是那两条黑色的线段。

总结一下，我的解释就是：大脑错认为我们在观察一个3D图案，并且按照经验认为远处的物体小，近处的物体大，因此分别对于感知到的刺激进行了心理上的权重更改，所以看起来，实际相等的线段在心理上感觉就不相等了。

我的实验是这样设计的：

 改变这两条黑色线段的倾斜角度，让上方的开口更大，下方的开口更小，这样的话，上方的红色线段离我们更远，下方的红色线段离我们更近，但是透视效果相比原图已经减弱了很多。之后，逐步改变角度，让两条黑色线段平行，这样就完全没有了透视的信息，两条红色线段和我们的距离相等。为了好玩，我还打算让这个图形倒过来，也就是说，上方的开口更大，下方的开口更小，这看上去上方的红色线段离我们更近，下方的红色线段离我们更远。

我对于实验的预期是：不论如何改变实验条件，离我们更近的总是被认为较长，离我们更远的总是被认为较短，并且，随着透视强度的增加，这种错觉现象会更加严重。


![](tmp2.png)
如上图所示，A1和A5的透视强度最强，接着是A2和A4，A3完全没有透视效果。我们预期效应在A1、A5上比A2、A4上要大，而在A3上则完全没有。

实验大概这样进行，给定红色的线段P，让被试在不同的位置（C1、C2、C3、F1、F2、F3）画一条线段，让其和红色的线段相等。被试不能依靠任何外界参考并且需要快速的画线。我们预计，越靠近红色线段，这种效应会越轻微，越远离红色线段，这种效应会越大。关于距离和效应的关系，我认为其意义在于，其可以说明远近对于效应影响的程度。

不论是竖线造成的透视效果还是线段距离红色线段的距离造成的效果，我认为都是感知远处和近处的差异造成的。如果一条线段，我们感知其离我们越远，那么其就越小，如果我们感知其离我们越近，那么其就越大。以上对于透视效果的操纵和对于距离的操纵可以说明这一点：我们改变了对于距离的感知，然后造成了大脑中根据经验进行的心理判断——被认为在远处的线条被主观的缩短了，被认为在近处的线条被主观的加长了。

这对于视觉感知概念有什么影响？

我始终坚持认为Dale Purves 教授的观点是正确的，这不仅仅因为我自己的生物学背景和对于进化论和进化心理学的了解，而是因为，这是从科学出发唯一说得通的解释。这种解释虽然不是预测性质的，但是其几乎可以解释所有的视觉错觉问题，这是非常迷人的事情。我认为一个优秀的理论就应该这样。如果这种说法是正确的，我们就不能再说“视觉错觉”这个术语了，因为这些“错觉”都是我们的视觉学习系统根据之前的经验得到的参数所做出的最大概率可能的决策。实际上，这些错觉很少会出现，并且我们的视觉系统也是高效率的，虽然带有一些缺陷，比如不能够完全得到物理世界的信息。

总的来说，我现在对于知觉的理解更加深刻了。之前我一直按照教科书上所讲的，外部世界的特征对应我们所感知到的特征，大脑所做的只不过是一种信号的转换。唯一的问题就是我们以什么样的方式来对外部世界的特征进行提取？但现在我认为特征检测的理论更多的贡献是，那群科学家发现了神经元存在很多种类并且对不同的我们已经感受到的刺激有不同的反应。我认为Purves的这些工作发生在外界物理刺激到我们所感受到的刺激的这段过程中。之后才是特征检测理论发挥作用的时候。

我认为不可逆问题的解决就是整个知觉领域面临的最大的问题，而围绕这个问题对知觉进行定义会是一个不错的选择。这就是我在学习玩这门课程后对于知觉新的看法。

顺便说一句，我认为有可能，我们可以通过神经网络模型来解决这个不可逆问题，训练大量的样本，然后模拟人脑的作用方式，找打一组参数，然后猜测大脑如何应对不可逆问题的。
