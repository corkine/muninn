### 3.1 冯诺依曼计算机概要

**所有的计算机，其基本的原理如下：**

- 二进制表示数据
- 布尔代数进行运算
- 电路实现布尔运算


当代计算机本质上都是以冯诺依曼（存储程序计算机）为基础而构建的，其包括I/O，控制器、运算器和存储器几个主要部分，高级语言通过调用CPU指令来操作内存的数据，进行计算。不同的是，计算机的制造材料一直在变化，从最开始的电子管到现在的晶体管，到未来的量子实现。

**步骤**

在控制器指挥下从存储器取出命令，分析指令后得到计算命令和待操作的数。

从存储器上取出待操作的数，放到运算器，根据布尔代数计算出结果，输出到储存器或者输出设备。

**对照**

运算器、控制器和存储器的一大部分都被集成在CPU中。



### 3.2 存储器原理和类型

bit 位， 8 bit = 1 byte 字节

byte 字节 是可以控制的最小字节

1KB = 1024 BYTE

1MB = 1024 KB

1GB = 1024 MB

1GB = 1024 GB

1TB = 1024 GB

1YB = 1024 PB

存储器包含 寄存器、高速缓存、内存和外部存储。

**寄存器**

寄存器用来存储特定操作数和结果，高速缓存用来做数据缓冲。内存是CPU想放但是放不下的数据。


![](cs03.png)

**高速缓存**

高速缓存分内外部，内部在于CPU内部，叫做Internal Cache，外部位于主板，叫做External Cache，纳秒内读写一次，容量在nK-nM。

**内存**

工作频率和前台总线频率相同，比CPU低，一般用于存放CPU运算数据，或者需要和外部存储交换的数据，断电丢失。

![](cs04.png)


CPU工作会从近到远寻找数据，找到后拉到自己附近，这样尽可能快的运行。

CPU具有数据访问的局部性，如果一个内存地址正在被使用，那么它很有可能再次被使用。（时间局部性）。在最近的将来可能用到的信息可能和当前使用的信息相邻（空间局部性）



**RAM 随机访问存储**

Dynamic RAM 可随机存取，但必须周期性刷新以保存存储内容。

DDR (Double Date Rate Synachronsous Dynamic Random Accessable Memory)双数据输出同步动态存储器

**ROM 只读存储**

只读存储器 - PROM 可编程写入ROM - EEPROM 可擦除重写ROM - Flash EEPROM 快速可擦除编程只读存储器

**32bit CPU**只能够寻找 $2^{32}$ 位的地址，这个数据约等于 4GB.

**赋值表达式及其优先级**

运算符优先级：括号大于乘除大于加减大于等不等大于赋值。其中赋值从右向左，而其余则从左到右？不一定。

`b = a == 3` b 的值为1，运算符 == 优先于 =。

`b = a = 3` b 的值为3，因为 a = 3 返回值——3。注意，不能使用 int a = b = 1;

`a+ = a- = a×a` 如果a为12，则 a- = 12 - 12×12 = -132，a+ = -132 - 132 = -264



**算数运算**

算数运算运算符具有不同的优先级，并且数值有不同的类型，各类型运算的时候要照顾优先级问题，也要进行类型的截断或升级。

![](cs17.png)

可以从最低优先级的位置开始将表达式截断，之后分别进行计算，最后根据各部分类型合在一起。

**自增自减运算符**

![](cs18.png)

负号和i加加相比，i加加优先级高，因此输出-3，i为4.

负号和加加i相比平级，但是遵循自右向左顺序，因此输出-4，i为4.

由于C规定自增自减运算符只能用于变量而不能用于表达式，所以下面两个是错误的。

**关系运算符**

小于、大于、大于等于、小于等于 的优先级相同

== 和 != 的优先级低于上述四者。

**运算符的优先级**

![](cs19.png)

使用剪刀法从较低级别运算符开始剪断，直到能够运算为止。


**逻辑运算符**

! 运算优先级 高于 && 高于 ||

`!a&&b||c --> ((!a)&&b)||c`

![](cs20.png)

```c
int main(void){
    int a = 0, b = 0;
    a = 5 > 3 && 2 || 8 < 4 - (b = 3);
    cout << a << " " << b;
    return 0;
}
```

其中 b 的值为0，a 的值为1，根据剪刀法来判断。需要注意，程序不会去计算（赋值）b。因为逻辑或运算左边为1，直接返回了结果。

**对于逻辑运算，非常重要的一点是，只有当执行下一个逻辑运算符才能够求出来结果的时候，才会进行运算。**

**逗号运算符**

逗号运算符用来连接表达式，一般用在 for 循环的第一个判别条件。

`a = 3 * 5, a * 4`

逗号在所有运算符中级别最低，甚至低于赋值运算，所以计算结果为 15*4 = 60

`x = (a=3,6*3)` x = 18 , 因为3被18冲掉了

`x = a = 3,6*3` x = 3

**条件运算符**

`exp1 ? exp2 : exp3`

等价于 `if exp1: exp2; else: exp3;`(Python)

可以这样写： `max = (a>b) ? a : b` max 等于 a、b中最大的数

**强制类型转换**

和常量类型规定类似：`1.234L`,可以对于常量和变量使用强制类型转换，这是一个表达式，这意味着我们仅仅是将这些量的值进行了转换，而这些量的类型本身没有变化。

`char a[10] = {'a','b','c'}` --> 没有初始化的元素变成 '\0'，这是 ASCI 码中的第一个值。

`char c[] = "China"` --> C h i n a \0

`char c[] = {'C','h','i','n','a'}` --> C h i n a

区分反斜杠字符数组、字符数组和字符串。字符串总是字符数组加上一个`\0`。

不能用赋值语句来将字符串赋值给一个字符数组：

`str1[] = "China" ; str1 = "China"; str2 = str1;` 均错误，如果需要赋值，需要通过循环去除\0，然后赋值元素，之后再添加\0.

**从CIN获取单个字符输入**


<u>get是一个通用函数，可以用来接受1个或多个字符，需要定义限制长度和终止符。对其进行简化的是getchar()和getline()，前者接受1个字符，后者默认终止符为\n。</u>

```c
char c;
while (cin >> c)
    cout << c;
```

当按下回车或者换行符时，才从输入缓冲区获得输入内容，因此输入`abc def g\n`会执行 `abcdefg` 程序会自动跳过空格，没有\0标识符。

```c
char c;
while ((c = cin.get()) != EOF)
    cout << c;

//等同于
while (cin.get(c)) //获取输入到c，如果输入EOF，则返回0
    cout << c;
```

这种方式不会跳过空格和回车，可以保留字符串样子。一直到输入EOF符号则结束输入。

```c
char c;
while (c = getchar()) //不跳过任何字符，包括ctrl+z
    cout << c;
```

**COUT输出字符数组**

```c
char a[3] = {'a','b','c'};
char b[] = "China";

cout << a << endl; //abc乱码中...
cout << b << endl; //China
```

COUT 输出字符数组会当期遇到第一个\0后停止输出，如果没有\0，则后面乱码。


**从CIN中获取多个字符输入**

```c
char str[10];
while (cin >> str) cout << str; 
//读取字符串，当遇到空格或者换行停止读取当前字符串。

cin.get(ch,10,'\n') 
//读取9个字符长度字符串，遇到\n则终止，读取到ch中，返回0或者1.
    
cin.getline(ch,10,'\n')
//类似于get，不过其遇到终止字符时，将缓冲区指针移动到终止字符之后
//而get则会将指针停止在终止处，而不会跳过终止字符。
```

getline 非常常用，因为通常将\n作为换行符，直接跳过了这个符号，则得到了一行的内容。

`cin` 使用输入缓冲区作为输入信息的来源，当按下\n或者缓冲区满的时候，缓冲区数据会被输送给程序。而在此之前，所有的输入都没有接触到程序。

上述程序中需要注意的是，cin >> n读取（刷新缓冲区一次）后指针停在了换行符上，因此在for循环中少执行了一次（有一次在读取这个\n的换行符）。需要使用 cin >> n 后添加一句 cin.get()，就可以避免for循环读取那个换行符。
