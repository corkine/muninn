# 函数

函数是一种封装，这种封装接受一些数据，然后经过某些运算后传出一些数据。毋庸置疑，函数的作用在于：对于自己：提供了一种代码复用的形式。对于他人，提供了一种不需要知道具体细节和实现就可以使用的黑箱。

函数包括其参数、返回值类型、函数名称和函数体组成。和变量一样，函数需要定义，称之为函数原型。函数原型扩展了在不同文件中调用函数的能力。

## 函数的原型

函数原型包括 返回值类型、函数名称、参数或参数类型三部分构成。

`float pow(float a, int b);` 其可以简写成 `float pow(float, int);`，省略形式参数的名称。

函数原型可以放置在专有的文件中，称之为头文件，比如 `math.h`，而使用的时候，只需要告诉预编译器  `#include <math.h>`即可。函数原型不是必须的，对于较小的文件来说，只要函数的调用在函数的定义之后，就可以不写函数原型，或者说，现在的函数原型就是函数的定义。

## 函数定义和调用

函数定义包含了完整的函数结构，当调用函数时，程序会通过函数原型找到函数定义，然后去执行。函数定义类似于：

```c
float pow(float a, int b){
    for (int i = 0; i < b;i++)
    res *= a * b;
    return res;
}

pow(10.3,20) //实参，是一些值
```

函数体中的a和b被称作形式参数，形式参数是一组变量，在调用时值被传递并且复制给形式参数。传递给形式参数的值称之为实际参数（实参），实参可以是任何能够生成值得东西，包括常量、变量以及表达式，实参总是将值传给形参，传递的总是值而不是变量，也就是说，传递过程是数据的复制过程。

## 参数传递过程

参数传递的机制大概如下：

C程序会单独开辟一片内存空间用来放置给定的参数（实参），称之为Stack堆栈，然后根据形参的数据要求从堆栈中取出对应长度的数据，比如第一个形参要求float类型，C就从内存中取出float类型占的字节长度，如果碰巧传递过来的值是char类型，那么C就会很不智能的跨过边界，其取得值将包括这个char以及其堆栈后紧邻的部分数据，这回导致错误，这种错误在实参部分缺失或者实参和形参不匹配的情况下会出现。

ANSI C 使用函数原型修复了不匹配数据类型的问题。

## 变量作用域

函数有变量的作用域问题。所有包含在函数内部的变量作用空间仅仅在其内部，而不会影响到外部的变量，除非是指针指向的地址。



# 指针

## 变量三要素

变量的三要素是变量的名称、其在内存中所表示的的值以及其在内存所占据的地址。

在Python这类解释型的语言中，很少关注变量的地址，所有变量使用结束后会被自动垃圾回收。而对于C而言，所有内存中的地址都可以手动管理。

代码或者数据，在内存中的地址，称之为指针。

比如：c = 27, c 在 内存中以 0041FE24 地址开始，占据int字节个内存块，那么就说 c 的地址是 0041FE24, 使用 &c 可以获取 c 对应的地址，使用 %p 来打印这个地址。

反过来，对于一个地址而言，可以获取其值。比如， *&c 就是 c 的地址对应的值， * 符号用来取地址的值。

## 指针变量

有了指针，那么我们可以让某个变量的值等于这个指针，比如声明 ptr = &c, 它的意思是，ptr这个变量的值是一个地址，因此使用 *ptr 可以获取 c 的值，也就是 ptr 的值（一个地址） 的值。

需要注意的是， ptr 是一个变量，其在内存中也有对应的地址，&ptr 会返回 ptr 这变量对应的地址， 这个地址和 ptr 的值表示的地址不同，这点需要注意。也就是说： &ptr 不等于 &c。

## 指针变量的定义

指针变量的定义采用如下形式： `int * ptr;` 其中 * 表示指针的类型，而int则表示指针的基类，也就是指针的值指代地址的数据类型的长度。这样的定义的原因很容易明白，因为这个地址仅仅是变量开始的地址位置，如果不说明这个变量的长度，那么指针就无法正确的知道其所指代地址的变量在何处结束。

良好的编程习惯总是在声明指针变量时赋予其一个专用的值 NULL。 `char * ptr = NULL;` 如果不给予初始值，那么一旦进行指针变量的赋值操作，谁也不知道这个指针刚开始指向何处，何处的地址又会被修改，这会造成巨大的损害。

## 指针变量的赋值

指针变量可以被赋值，如下：`ptr = &c`

需要注意，其中 *ptr 指的是 ptr 表示地址的值，等于 c；

&ptr 指的是 ptr 变量的地址，和 c 的地址不同，不等于 &c；

ptr 本身指的是一个地址，其和 &ptr 也不相等。

**注意：** *ptr 中保存的到底是 c 还是 c 的值呢？ 应该是变量c，因为其被赋值的时候赋予的是c这个符号的地址，这个地址很有可能和c真实数据存储的位置不同，其和c的数据无关，如果c更改，那么这个值也会被更改，因为其指向的是c这个符号。

## 指针变量优先级

后置++ -- 优先于 前置 ++ --，！，* 以及 &

但是根据就近原则，下列公式成立：

`&*ptr = &(*ptr)`

`*&ptr = *(&ptr)`

## ptr ++

`ptr++` 指的是什么意思呢？其指的是下一个地址，注意这里的下一个地址不是说指的是相邻的内存字节，而是指的是跨越其类型长度字节后的下一个字节，比如 `a[] = {1,2,3}, *ptr = a`, 那么 `ptr++` 之后等于 `ptr` 等于 `a[1]`。这里跨越了一个int的长度（4）。


# 一维数组和指针

定义： `int a[] = {1,2,3,4,5}`,那么 `a, &a &a[0]` 的值完全相等，实际上数组的值就相当于数组第一个元素的地址。这个结论非常重要，这意味着，所有数组操作都可以被替换成指针操作。

## 数组访问

`*(ptr+i) = *(a+i) = a[i] = ptr[i]`

如上所示，方括号取下标的操作使用数组变量也可以（其实a等价于ptr）。更重要的是，使用(ptr+1)或者(a+1)这种形式可以取数组的下一个地址，然后使用*运算符来获取其内容，这和方括号取下标也是等价的。

这里有一个小问题，对于指针来说，其操作更不安全，因为对于下标操作，a[1000]会报错，而`*(a+1000)`则不会，并且很有可能能够返回结果。此外，`*ptr++` 可以用来遍历数组，但是 `*a++`就不可以。这也说明了在某些情况下a并不能完全代替ptr。

**a++ 毫无意义，并且会报错**

# 二维数组和指针

`a[3][4]` 可以看作是 `a[0], a[1], a[2]` 组成的三个部分的一维数组，而对于其各自部分，又可以分成 `a[0][0]...a[0][3]`。在定义指针的时候，不能采用一维数组那种定义方式，现在的基类型指的是这些各自的部分，其中每部分包含了长度为4的数组。 

如下图所示：
```
变量      值
a        &a[0]      a+1    -> &a[1]
a[0]     &a[0][0]   a[0]+1 -> &a[1][0]
a[0][0]  some value
a[0][1]  some value
a[0][2]  some value
a[1]     &a[1][0]
a[1][0]  some value
a[1][1]  some value
...
```

## 指针声明

`int (*ptr)[4]` 采用这种定义方式很明显的可以看出，首先计算括号里的数值，*ptr创建了一个指针，然后这个指针指向 一个长度为4，每个元素为int的数组。 而 `int *ptr[4]` 则会先创建一个长度为4的ptr数组，然后创建一把指针。

## 指针和下标

*p 表示的是对应第一维数组的第一部分，也就是 a[0]

*(p+i) 表示的是 a[0+i]

*p 是一个地址，`**p` 也是一个地址，这意味着什么？你可以看作：变量p处放置了一个链接，而这个链接又指向了一个链接，而这个二级链接就是a[0][0]

因此，采用 `*(*(p+i)+j)` 可以访问 a[i][j]。

# 数组指针的级别和运算

## 运算规则

对于数组指针的运算，比如 a[] = {1,2,3}, a+1 会得到 a[1] 的地址。以下是三要素：

- 对于数组名而言，其值永远是指向第一个元素的指针

 - 对于一维数组，a 的值就是 `a[0]` 的指针/地址
 
 - 对于二维数组`a[][]`, a 的值就是 `a[0]` 的指针/地址， `a[]` 的值就是 `a[][0]` 的指针/地址。需要注意，二维数组并非是拉长的一维数组，二维数组的定义是基于其所包含一维数组来的。
 
 
- 对于 * 操作而言，总是向下降一级

 - 对于一维数组， `*a` 的结果是 a[0] 的地址变成了 a[0] 的值，从地址变成值是一种降级。
 
 - 对于二维数组， `*a[2]` 的结果是由 第三个数组的指针/地址 变成了 第三个数组中第一个元素的地址（也就是 `&a[2][0]`，解释：a是第一个数组的指针，所以`a[2]`是第三个数组的指针，第三个数组的值根据定义，等于`&a[2][0]`,也就是第一个元素的地址，因此，第三个数组的指针取`*`得到的是第三个数组第一个元素的值，即 `*a[2] = *&a[2][0] = a[2][0]`。
 
 
- 对于 & 操作而言，总是向上升一级


## 数组级别

根据以上内容，对于二维数组，显然有：

```c
a = &a[0]
a[0] = &a[0][0]
a[0] = *a
a[0][0] = *a[0] = **a

//对于二维数组，其级别为：

&a // +1可以跨越整个二维数组
  a // +1可以跨越一个一维数组
    a[0] // +1可以跨越一维数组的一个元素
      a[0][0] // +1没有含义，这里已经是值，而不是地址了

```

## &a = a[0]

【注意&a】作为特例，对于一维数组a，&a表示升一级，但是其值还等于a，都是代表第一个元素的地址。但是区别在于 a+1 和 &a+1，前者跨越了一个数组元素，后者则跨越了整个数组，这可以看作升级带来的不同。

对于二维数组而言，&a[0] = a[0]显然也是这个例子的说明，因为 a = &a[0], a[0] = &a[0][0], 为了正常运算， 对于数组名而言， &a[0] 总是等于 a[0]，这样 a = &a[0] = a[0] = &a[0][0]。这是需要注意的点，其中中间的等式是由这个规定而来。

# 字符串和指针

字符串就是一个类型为char的数组，结尾以'\0'结尾。一个神奇的事情在于可以将字符串赋值给一个指针，比如 ptr = "hello"，这个指针指向不可修改的字符串数组的地址，使用 %p 可以查看首字母地址，使用 %s 可以直接生成所有字符串（某种宏？），使用`*p`可以返回h这个字母，和数组指针类似。

需要注意，在cpp环境下的cout输出ptr会直接被转换成 %s 而输出字符串，而不是输出一个地址。在c下不存在这种问题。

# 函数和指针

函数和指针关系密切，其中函数的参数可以是指针，这样函数就能进行原地修改而不是复制一份数据，效率高了很多，不过安全性也下降了很多。此外，函数的返回值也可以是指针，不过要注意指针的有效性问题。

## 参数作为指针

```c
int change(int *a,int *b)
{
    int temp;
    if (*a < *b){
        temp = *a;
        *a = *b;
        *b = temp;
    }
    return 0;
}
```

**二维数组相关转换与定义**

参数作为指针的时候，在声明参数的时候需要进行指针的定义，如上所示。同时，这里可以进行复杂的定义，比如：

```c
int findmax(int a[][4])
{
    a++; //不是错误
    return *(*(a+1)+1);
}
int findmax(int (*a)[4])
```
这两者是等同的，并且前者总是转换成后者进行运行，这意味着你可以在第一个定义中调用 a++，这在非函数中是禁止并且编译器会报错的。


## 返回值作为指针

返回值作为指针的时候，定义如下，需要注意变量作用域问题。

```c
int *f1(void)
{
    int a = 233;
    return &a;
}
```

很显然的是，这个程序不能正确运行，因为这个指针，也就是函数返回的指针指向的是函数内部的值的地址，而这个值在函数运行结束后就消失了，因此其指针指向的地址无意义。

为了改变这种情况，使用静态局部变量即可。


## 静态局部变量和指针

```c
void f2(void)
{
    int a = 0;
    static b = 0;
    a++; b++;
    cout << a << ", " << b << endl;
}

int main(void)
{

    for (int i = 0; i < 10; i++){
        f2();
    }
    return 0;
}
```

这个程序而言，a在每次main调用f2函数时总是被重置，而b应为声明为静态局部变量，因此其值不会被垃圾回收，在每次f2()调用结束后继续存在，因此就进行了累加。b分别为0,1,2...9。函数一般的局部变量都是动态的，也就是会自动进行垃圾回收的类型。


# 指向地址常量的指针

指针权限太大了，因此可以采用 const 定义静态指针，比如 `int findmax(const int a[][4]); int findmax(const int *ptr)`，其作用在于，指向常量的指针不能更改其储存的值的地址。

比如：

```c
char *mystrcpy(char *a,const char *b)
{
    a = b;
    //b = a; //可以，指向常量的指针并非不能被赋值，只是不能更改其值对应地址的内容
    //*b = 233; //不可以
    return a;
}
```

这个函数保护了被复制的字符串不被更改。需要注意，指向常量的指针并非不能被赋值，只是其不能够更改其值对应地址的值，即 `*ptr` 不可赋值。

比如:

```c
const int a = 1;
const int b = 2;
int c = 3;
const int *p = &a;
*p = 2; //错误
p = &b; //正确
*p = 3; //错误
p = &c; //正确
*p = 233; //错误，p的const由指向符号常量的p定义所约束，和地址指向的值无关。
```
