# Week 3 独立概率和古典概率 [SP1.4]

## 1. 事件概率的独立判断

**定义**：$P(A\cap B)=P(A)·P(B)$

由条件概率公式：$P(A\cap B)=P(A | B)·P(B)$可得独立概率的第二个直观定义为：

**定义2**：$P(A|B)=P(A)$

**例子**

可以根据概率独立的定义来判断两个事件是否独立，以及这两个事件的条件概率。比如：

对于拖车和学生，如果学生是美女，那么请求不要拖车，被答应的概率会高一些，如何验证？

我们知道在所有同学中，是美女的概率是一定的，其次，拖车时，是否求情的概率也是一定的。所以，根据观察，得到：

$P(女生没有求情且车未放行)=0.85，P(女生没有求情且车放行)=0.05$
$P(女生求情且车未放行)=0.01，P(女生求情且车放行)=0.09$

根据定义，我们需要计算$P(放行)$和$P(女生求情)$，已知放行概率为0.2，女生求情概率由上可知，为0.1，那么，显然女生求情和放行不是独立事件。那么，$P(放行|女生求情)=0.09/0.1=0.9$，也就是说，如果是一个女生，且其求情的话，被放行的概率为0.9，远高于一般人放行的0.2。

> 需要注意，在实际观察中，我们不能获取条件概率，只能够获取一般事件的概率以及复合事件（cap）的概率，但是，我们可以根据cap以及一般概率推出来cup、条件概率，以及根据独立判断公式判断事件是否独立。

实际容易犯的错误是将$P(A\cap B)$看作$P（A|B）$

**多个事件独立**

多个事件独立意味着其中两两独立、三三独立、nn独立，需要计算$A^n_n次$

## 3. 古典概率

古典概率是一个**数数**的问题，其假设每个实验结果（outcome）（S中的每个元素）发生的几率相同，那么，一般的解决办法是：
    
$总的概率 = 出现的次数 * 相同的概率$
    
所以说，本质在于数数(计算某事件包含多少实验结果)。

### 3.1 基本原则

若某种实验有n种不同的结果，而另一种实验有m种不同的结果，那么，按顺序操作两实验，则有n*m种不同的结果。

（每种结果对应S中每个元素，其在古典概率中发生概率相等，因此只用计数即可）

**基本原则非常重要，你可以将其看作是概率树：n个分支，每个分支都有m个子分支，一共有m×n中结果**

### 3.2 基本判断

所有的对象是否可区分？

选中的对象是否可放回？

抽选的顺序是否有差异？

### 3.3 排列和组合

如果不可放回，有顺序差异，则是一个排列问题。

如果可放回，则是一个$n^k$的累积问题（又叫重复选取）（一种特殊的排列）。

如果不可放回，没有顺序差异，则是一个组合问题。


**Example**


若某种实验有 n 种不同结果，而另一种 实验有 m 种不同结果。若操作此两实验将有n·m 种 不同结果。

Ex: 下午茶有 5 种甜点的选择， 10 种饮料的选择。

共有多少种下午茶组合？ 5x10 = 50

**排列 (Permutation)**

Ex: 小美周末两日惯购物。小美常自明华园 三兄弟找人接送。若两日司机不可重复，问有多少种结果？

3x2 = 6

若有 n 异物，从中依序取出 k 物 共有多少种结果？

n! / (n-k)!

**重复选取 (Choose with Replacements)**

Ex: 小美周末两日惯购物。小美常自明华园 三兄弟找人接送。无耻小美竟敢不排除连续凹人两天， 问有多少种结果？

3x3 = 9

若有 n 异物，从中选取一物，每 次取完放回。依序选取 k 次，共有多少种 结果？

nᵏ

**组合 (Combination)**

Ex: 小美爱玩跳棋。小美常自明华园三兄弟 找两人下棋。问有多少种对战组合？

3x2 / 2 = 3

若有 n 异物，从中取出 k 物 共有多少种结果？

n! / ( (n-k)! · k! )

### 3.4 多项组合

组合中有一种特殊的情况称之为**多项组合**,比如：

Ex:费雯兄惯于网络八卦版上发废文。第一楼推文常有四类
–你妈知道你在发废文吗
–见此唉滴必嘘
–在五楼…
–妈！我在这！

•问：费雯兄发文𝟏𝟎次，一楼推文共有多少种组合？ $C^{4}_{10}$ or $10^4(√)$ 注意使用基本判断三维度

•问：有多少组合会看到𝟒次「你…」，𝟑次「见…」，𝟐次「在五楼…」，𝟏次「妈…」？

$C^{4}_{10}·C^{3}_{7}·C^{2}_{3}·C^{1}_{1}$

这是一个多项组合的问题，往10个空中填四次，第一次填入4个你...，有$C^{4}_{10}$种可能，第二次填入3个见...，有$C^{3}_{7}$种可能，以此类推。

需要注意，多项组合需要根据之前的情况进行考虑。

继前述费雯兄好发废文之例。若根据统计，费雯兄一楼推文不同
型态之出现机率为：

–𝑷(「你妈知道你在发废文吗」)=𝟎.𝟒

–𝑷(「见此唉滴必嘘」)=𝟎.𝟐

–𝑷(「在五楼…」)=𝟎.𝟏

–𝑷(「妈！我在这！」)=𝟎.𝟑

•问：若费雯兄发文𝟔次，会在一楼推文看到𝟐次「你…」，𝟐次「见…」，
𝟏次「在五楼…」，𝟏次「妈…」。这样的机率为？

反正概率相同，先随便求一个：比如，𝑷(你你见见五妈)（有顺序）

之后，个数×P就是概率：

𝑷(2你, 2见, 1五, 1妈)= $C^{2}_{6}·C^{2}_{4}·C^{1}_{2}·C^{1}_{1}$ * P


**总而言之，多项组合和之前的二项组合非常不同，不是从一个空间中选取某一个或者几个元素，而是将这些元素选取到大于2个的不同项目中。**选取的时候按照顺序考虑就好，将其看作是几个不同二项组合的独立实验，然后根据基本原则求其乘积。

### 3.5 概率树和古典概率

可以将概率树分为一些彼此独立的实验，根据古典概率基本原则，独立的实验，最终结果为其相乘的值。

比如，一个母亲要将5只铅笔分给三个孩子，对每个孩子而言，其可能不要铅笔，可能全都要。那么，可能的分配方案有几种？

将其分为几个独立部分：3个孩子没有铅笔×0个孩子组合铅笔+2个孩子没有铅笔×所有铅笔全给剩余1个孩子+1个孩子没有铅笔×2个孩子组合铅笔+0个孩子没有铅笔×3个孩子组合铅笔。

古典概率是一种特殊的概率树，其每个分支的概率均相等。我们可以将概率树切分成不同的单元来将复杂的问题简单化。

比如：柏拉图花园采玫瑰问题。一共有8朵玫瑰，其中柏拉图要采最漂亮的一朵。他只能够比较花朵得出漂亮与否的结论。他给自己定了一个规则：对于8朵玫瑰，前三个不要，从第四个起，如果其比前三个都漂亮，则要第4个，否则，继续前进，重复进行比较。

每次的决策受到之前决策的影响，柏拉图在每个位置得到最漂亮玫瑰的概率为：
（0 + 0 + 0 + 1 + 3/4 + 3/5 + 3/6 + 3/7 ） / 8 = 0.41
