## 经验之谈和数据之道

### 来自于经验的回答存在的问题

如果探讨“第一个孩子出生晚吗？”这种问题，在社区中间进行提问，我们会有以下几个难以解决的问题：

1、选择偏差：进入社区板块的人更有可能是第一个孩子出生晚或者对于这件事耿耿于怀的父母

2、样本不足：几个偶然样本不能说明一个结论

3、不准确：父母回忆的数据有可能带有偏差和夸大

4、确认偏差：对于观察者而言，其更愿意把那些和自己观点一致的东西看作证据，而忽视那些不一致的信息。

### 来自于统计数据的解决方法

为了解决以上的问题，我们可以使用数据和统计方法来回答问题，我们主要做的是：

1、收集和清洗数据

2、描述性统计

3、探索性数据分析：如果说数据可以被用来说明问题，往往我们会在这一步关注这部分数据

4、假设检验：如果发现可能的关联，我们要确认这种关系不是发生在随机水平上的

5、估计：最后一步是由部分推论到总体。

## 横向研究和纵向研究

### 横向研究

横向研究（Cross-sectional study）又被称之为横断面研究，区别于纵向研究（又称之为case study、longitudinal study）。其定义为对总体或者一个代表性子集进行在某个时间点的观察以得出可能性的关系的研究方法。

由于其选取一个时间横切面，所以这种研究只能用于描述数据而不能精确的用于因果推论。

**优势：**横向研究的优势是其花费较低，能够在研究刚开始的阶段快速提供关于某几个变量可能相关的证据支持，以备在后续开展正式研究。

**缺点：**横向研究的缺点有四：

横向研究一般为问卷调查，其只能够给出某个时间点的关于某些变量的描述信息，我们不能够：

1、缺乏原始数据。这些信息一般不是专门为我们研究目的设计的，比如人口数据，都经过了汇总，因此可能在不同汇总维度我们想要研究的变量关系存在显著差异。

2、数据不客观，可信度低。这些信息是观察到的，其设计并不严谨，因此不容易准确的得到关于变量关系的测量（可能更多带有被试主观偏见）

3、数据只能反应现在状况并且尚未经过重复测量。这些信息只能反应当下的状况，且基本上很难进行重复测量。尝试回忆过去会带来回忆的偏差，这导致我们无法建立时间上的联系和进行可能的因果推论。

4、数据并非为我们原本目标设计，这些数据大多数不包含我们希望控制的那些混淆两个变量关系的中介因素。

### 纵向研究

纵向研究（Longitudinal study，panel study，cohort study（panel study的一个子集，指的是包含相同特征的一组被试））指的是在一段时间内对于同一变量的多次观察。

*维度1*：纵向研究可以分为两个维度，其一为不包含控制的纯粹观察，在这种情况下，我们可以把横向研究和纵向研究看作是被试间研究和被试内研究，由于后者对于个体层面进行了反复观察，能够排除时间相关的变量影响，而我们认为一个人内部变异在大多数问题上小于外部变异，因此其数据经常被认为是比横向研究可靠的。但这种假设可能在一些研究中不成立。

*维度2*：其二为包含观察和类似于实验的控制。在第一种纯粹观察的情况下，纵向研究被认为可信度在横向研究和实验之间。而第二种的可信度几乎和实验类似。


## 第一步

通过对于数据进行的简单比较（探索性数据分析）所得到的结论通常反映了“直观效应”。但是我们尚不能因此而得出结论。

导致结论错误的原因有很多，有可能我们没有顾及所有的统计量以及对差异的尺度把握不好，有可能这种差异是随机产生的，还有可能我们所得出结论的数据来自于实验设置的错误，比如选择偏差等。

我们需要对此进行精确的控制以得出可信的结论。概率论和统计学为我们提供了一套标准、可信的检验办法。

## 一些术语

总体population、样本sample

按人头分类：队列cohort、被调查者respondent、记录recode

按变量分类：表table、字段field

原始数据row data、重编码recode（重编码意味着在分类变量中由于某些种类数据较少，人为扩大样本中此类数据的比例然后给予降权以提高采样准确性的数据处理方法）

比如说我们需要对于100：1的汉族人和藏族人分层抽样，抽取101人的样本，那么这个藏族人如果抽取一个，其不能够很好反应藏族人的平均结果，因此抽取10个藏族人，平均得到一个藏族人的数据，这样数据更准确。
