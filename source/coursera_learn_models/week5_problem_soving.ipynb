## Perspectives 视角

视角s指的是**对所有能够想到的解决方案的表述**。视角指的是各种对问题不同的看待角度。比如笛卡尔坐标系和轴坐标系，它们有不同的表述方法。一个人的视角可能包含其中之一，也有可能包含这两者。

视角决定了不同的看待问题的方式，这些不同的看待方式决定了采用不同视角看待一个问题，感受到的难易程度不同。

## Local Optima 局部最优

局部最优指的是不论向左或者向右看，都是退步/不能找到更好的解决方案的那些点。

一个好的视角包含更少的局部最优解（只有一个山峰），而一个坏的视角则包含更多的局部最优解（许多不同的山峰）。

换句话说，对于某个特定的视角，局部最优点指的是那些不转换视角就永远无法解决的问题。比如说，你将一个人描述为高大的、英俊的，而另一个人是矮小的、丑陋的，你无法在他们都作弊这件事情上给出这种视角的解决方案，除非对其进行重新描述——有道德的和无道德的。

## Global Optima 全局最优

越是差劲的模型越容易遇到这些棘手的问题，而一个好的模型（表示方法和看待角度）则是连贯的，通用的，比如 F = M·a ,几乎对于任何物体都适用（宏观），在这种情况下，也就是较少遇到局部最有问题的情况下，我们称之为富士山模型，它较少崎岖的路径能够让我们在寻找问题答案时不会陷入过多的局部最优解（不论向左还是向右都是下降），而是一直上升或者下降。

## Team 

团队是视角多样性的来源

团队的最大作用就是提供了**多样性**的视角。而这些局部最优的视角的交集，更有可能是正确问题的答案。换句话说，如果有一个理论有更少的局部最优，那么它一定在交集里（容纳和解决其他团队人员的问题）。

# Markov Model

## Define Markob Model

- States 有限状态
- Trans Probability 概率恒定

```
        M(T)  N(T)      Curent   Futrue
    
M(T+1)   .75  .05       p        p

N(T+1)   .25  .95       1-p      1-p

```

M在一次转换周期中，转换为N的概率为0.25，N的转换概率为0.05。

概率矩阵乘以向量[p 1-p]'得到T+1时刻的概率，当T+1时刻概率和T时刻概率相等时，这时候称之为统计平衡。

## Markov Convergence Theorem

转移概率决定了一个均衡，并且此均衡唯一。这是否意味着中途干预没有必要？或许不是，因为即便马尔科夫过程成立，那么这个干预的平衡可能需要很长的时间。并且，马尔科夫过程中有一个条件是，可以从一个状态自由的转移到另一个状态，这个条件可能不成立。比如说自由国家和非自由国家的转换。
