# Modeling People's Mind

2018-05-19

有三种不同对人进行建模的模型，它们分别是：

## 1、Diversity Purposeful Actors - Rational Actor Model

### 概要

理性人模型，此模型主要内容为，假设人们的行为是有**明确目的（Purpose）**的，并且人们总是围绕着这个目的进行不断的**优化行为（Optimize）**，以获取**最大收益（Outcome）**。

基于理性的模型探讨的情况是——当人们精打细算，或者进行重大决策是，进行选择的情况。在这种模型中，决策过程非常重要，非常复杂。聚合过程也不简单，不过，相比较基于规则的模型的复杂聚合，还算简单。

### 举例

对于企业来说，最大收益是利润最大；对于个人来说，最大收益意味着最大效用。

还需要注意一点，理性不意味着自私。实际上，我们可以**理性的利他**。这取决于怎么看待一件事的最大效用。

### 分类

对于理性人模型，我们有两种不同的处理办法：

其一：决策Decision，决策是无关他人选择的选择。在上一节讲到了两种不同的决策思路。

其二：博弈Game，博弈论是一种特殊的决策，我们决策的结果取决于他人的决策结果。博弈也有相应的决策思路，通常是博弈Game Tree.比如，常见的博弈模型为以牙还牙，或者触发冰冻。

### 适用

对于群体决策而言，Group Decision使用理性人模型可以很好的解决，考虑群体最大化利益往往是一个群体决策的最优选择。

对于大量重复决策而言，Repeated Decision使用理性模型再好不过了，因为这是由理性模型的定义得到的。当我们进行一个决策的时候，可能选择的参数是不准确的（非最优），但是当对于一个问题有多次机会的时候，我们就可以调整参数或者最优化选择，这自然会得到最优的收益。

对于高风险决策而言，理性人模型的优势主要在于复杂的步骤和复杂性决策分解过程，这在很大程度上保证了结果最优。



## 2、Behavioral Model

### 概要

这是一个基于理性人模型的一个模型，不过基于“观察”，并且考虑到了一些心理层面的因素。

理性人模型在经济学中被大量使用，但是往往会观察到一些经常性的决策偏差，这些偏差是由生物学意义上进化的最佳选择，但并不是最大收益和理性的。行为模型主要探讨这些决策偏差在理性人模型中的影响。

基于行为的模型探讨的情况是——当人们的注意力被占据，而需要快速进行选择，或者选择的代价不太大的时候，快速利用心理机制进行选择的情况。

### 常见误差

常见的行为误差有：

#### Prospect thery 前景/期望理论

这个理论认为，我们对于收益有着风险最小的偏好，对于损失有着冒险的偏好，这可能是由进化而带来的心理机制。

比如：两个选择，A：得到400元；B：50%概率得到0元，50%概率得到2000元。

#### Hyperbolic Discounting 双曲贴现

比如，两个选择，A：立马得到100元，B：100天后得到105元。

比如，连个选择，A：100天后得到100元，B：105天后得到105元。

双曲贴现意味着我们对于近期折现的偏好。比如，你面前有块蛋糕，那么即便你想要减肥，但是也很难面对这种诱惑。

#### Status Quo Bias 现状保持偏差

比如，复选框，选中表示同意捐赠vs选中表示不同意捐赠。

现状保持偏差非常的常见，比如在消费一件没用过的东西时，在考虑是否续费时，等等。

#### Base Rate Bias 参照偏差

这个偏差是由注意力资源有限的情况下，由启动效应决定的。比如，在回答两个问题，随机最大，那么下一题有可能重复上一题的选项。或者很接近。此偏差对于视觉错觉，比如对某件事情的感知依赖于其背景 这样的结论非常的类似。

参照偏差依赖于我们当前的条件和背景，我们会常常在这样熟悉的情境下做出熟悉的（也可能是经常错误的）选择。我们的理由总是唯一的，以前就是这样做的，现在为什么不呢？在这种情况下，我们参照的标准是自己。

此外，对于一个较小发生的事件，我们经常低估它的可能性。但是，事件往往在某些小概率的条件中有大概率发生，即小概率条件的小概率事件其实有很大概率发生。在这种情况下，我们参照的标准是此事件独立发生的概率。

**我们所做出的选择总是相对的：好，是相对的，坏，是相对的。那么关键就在于所比较和参照的对象。我们对于这个对象的改变总是缺少必要的灵敏度，这就造成了参照偏差。**

### 关于这个模型

我更倾向于将这个模型看作是理性人模型的一个Add-on，但是，这些误差在定量的角度难以进行计算和评估，所以在这里单独列出。实际上，**我个人推荐在进行选择的时候考虑使用理性人模型，并且时刻考虑在不同的决策树分支，考虑进去这些误差所带来的概率变化。**

## 3、Rule Based Model

像是谢林分离模型，人们根据环境的变量进行独立决策。或者是起立鼓掌模型，人们根据环境中的变量决定是否模仿别人站立鼓掌，在对这些事情进行建模的时候，**决策发生在一瞬间，只有简单的规则和复杂的聚合过程。**

### 分类

基于规则的模型可以分为四大类：


Fixed - Common Decision、Fixed - Game Decision、Adaptive - Common Decision、Adaptive - Game Decision

其中Fixed指的是固定的模型，只要模型确立，就不再调整，这在某些情况下是方便的，比如快速的反应和决策。

Adaptive指的是动态的模型，对于很多重要情况，能够有一个可伸缩模型还是必要的。

FD的例子是随机模型，这是一个完全自我决策的模型。

FG的例子比如以牙还牙或者公平分配，这是一个静态的多人共同决策的模型。

AD的例子是Gradient梯度下降，比如我们在进行决策的时候进行梯度下降寻找最优解，不断有最小的Cost Function。

AG的例子是模仿Mimicry，这是一个动态的，多人决策的模型。

### 说明

简单的规则容易被他人利用，比如说讨价还价，当别人知道你的策略是比预期的高/低，他就可以针对性的进行反应。

## 4、三种模型的选择

这三种模型各有偏好，都非常常用。对于社会行为，经常性的我们使用基于规则的模型。对于需要快速反应的行为，我们使用基于行为的模型。对于需要精细打算的行为，我们使用基于理性人的模型。

通常来说，**我们可以对于一件事进行这三种行为的建模**，如果它们的结果相差不太大，那么无所谓，否则，我们就知道这件事的重要特点：比如，当基于行为的模型更优，那么我们就可以说，这件事的最大特征是：需要快速决策，并且不太重要。

对于理性模型而言，需要提的点是，如果将现实看作是一系列事件组成的合集，那么理性模型代表着出现频率最高的那组事件。当重复的事件在时空中频繁出现，那么，它总会向理性模型的那组事件偏斜。因此，理性模型虽然在实际使用中不太准确，但是这确实代表了在我们所知道的信息中进行最佳反应并拥有最大可能性的最佳选择。

# 线性和分类模型

这里的分类模型不是Logistic回归，而是指的这样一种模型：

对于一件复杂的事情，如何将其分解并且分为几种不同的类型是最优的选择？

如果说，将其分解为三种不同的类别，而某一类别占据的比重最大，我们就可以用这一部分来概括整个事情的内容（贴标签）。

这和PCA降维，Cluser很像。

计算方法和评价方法大概为：

$R^2 = 将其分为不同类别后各类别的方差之和/未分类时的总方差$

而线性模型，则就是线性回归。其$R^2 = 1- SSR/SST$。

线性和分类模型可以看作是在框架下的开始部分，对于数据的理解以获得信息，通常是何者具有更高的权重，然后进行决策的过程。

在这种思路下，数据驱动的机器学习的问题就很明显，只有数据处理的过程，而没有进一步的理解因果关系，进行决策、聚合的过程，所以被看作是残缺的。
